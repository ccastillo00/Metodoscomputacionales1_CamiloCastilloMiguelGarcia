\begin{equation}
Ax = b
\end{equation}

donde $A$ es una matriz triangular inferior y $b$ es un vector. El objetivo es encontrar el vector $x$ que satisface la ecuación anterior.

Podemos descomponer la matriz $A$ en una matriz diagonal $D$ y una matriz triangular inferior estricta $L$ como:

\begin{equation}
A = L + D
\end{equation}

Reemplazando $A$ en la ecuación original, tenemos:

\begin{equation}
(L + D)x = b
\end{equation}

Despejando $Lx$, obtenemos:

\begin{equation}
Lx = b - Dx
\end{equation}

La sustitución hacia adelante se realiza para resolver el sistema $Lx = b - Dx$, que es una matriz triangular inferior.

Comenzamos resolviendo la primera ecuación, que es de la forma $L_{11}x_1 = b_1 - d_{11}x_1$. Despejando $x_1$, obtenemos:

\begin{equation}
x_1 = \frac{b_1 - d_{11}x_1}{L_{11}}
\end{equation}

Podemos reorganizar esta ecuación para obtener:

\begin{equation}
x_1 = \frac{b_1 - \sum_{j=0}^{i-1} A_{ij}x_j}{A_{ii}}
\end{equation}

donde $i=1$, $j$ es la variable de la sumatoria, y $x_0, x_1, \dots, x_{i-1}$ son los valores de $x$ que ya hemos calculado.

Ahora, para la segunda ecuación $L_{21}x_1 + L_{22}x_2 = b_2 - d_{22}x_2$, despejamos $x_2$ para obtener:

\begin{equation}
x_2 = \frac{b_2 - d_{22}x_2 - L_{21}x_1}{L_{22}}
\end{equation}

Podemos reorganizar esta ecuación para obtener:

\begin{equation}
x_2 = \frac{b_2 - \sum_{j=0}^{i-1} A_{ij}x_j}{A_{ii}}
\end{equation}

donde $i=2$, $j$ es la variable de la sumatoria, y $x_0, x_1, \dots, x_{i-1}$ son los valores de $x$ que ya hemos calculado.

Continuando de esta manera para todas las ecuaciones del sistema, podemos obtener la siguiente fórmula general para calcular $x_i$:

\begin{equation}
x_i = \frac{b_i - \sum_{j=0}^{i-1} A_{ij}x_j}{A_{ii}}
\end{equation}
