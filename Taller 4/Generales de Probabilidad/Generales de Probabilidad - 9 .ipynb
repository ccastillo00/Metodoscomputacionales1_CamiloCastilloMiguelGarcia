```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
p1 = 0.5 
PA = 6 * p1**2 * (1-p1)**2
print(f'La probabilidad de obtener dos caras y dos sellos es: {PA}')
```

```python
p1 = np.linspace(0.1, 0.9, 100)
p2 = np.linspace(0.1, 0.5, 100)
P1, P2 = np.meshgrid(p1, p2)
PA = 6 * P1**2 * (1-P1)**2 * P2**2 * (1-P2)**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(P1, P2, PA)
ax.set_xlabel('p1')
ax.set_ylabel('p2')
ax.set_zlabel('P(A)')
plt.show()
```

```python
p1 = 0.5
p2 = 0.5
PA = 6 * p1**2 * (1-p1)**2 * p2**2 * (1-p2)**2
print(f'La probabilidad de obtener dos caras y dos sellos es: {PA}')
```
