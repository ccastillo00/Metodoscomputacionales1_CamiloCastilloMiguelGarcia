```python
import numpy as np
import matplotlib.pyplot as plt
```

\begin{equation}
\begin{aligned}
P(CSCS) &= p \times (1-p) \times p \times (1-p) = p^2(1-p)^2 \
P(CSSC) &= p \times (1-p) \times (1-p) \times p = p^2(1-p)^2 \
P(SCCS) &= (1-p) \times p \times p \times (1-p) = p^2(1-p)^2 \
P(SCS) &= (1-p) \times p \times (1-p) \times p = p^2(1-p)^2 \
P(SSCC) &= (1-p) \times (1-p) \times p \times p = p^2(1-p)^2 \
\end{aligned}
\end{equation}

La probabilidad total de obtener dos caras y dos sellos es la suma de las probabilidades de los seis caminos:

\begin{equation}
P(A) = P(CCSS) + P(CSCS) + P(CSSC) + P(SCCS) + P(SCS) + P(SSCC) = 6 \times p^2(1-p)^2
\end{equation}

Para obtener el resultado de $\frac{8}{3}$, debemos tener:

\begin{equation}
6 \times p^2(1-p)^2 = \frac{8}{3}
\end{equation}

Dado que tenemos las dos primeras monedas truncadas, podemos calcular las probabilidades para los seis caminos diferentes. Consideremos lo siguiente:

\begin{align*}
\mathbb{P}(CCSS) &= p_{1} \cdot p_{2} \cdot 0.5^{2} = p_{1} \cdot p_{2} \cdot 0.25 \
\mathbb{P}(CSCS) &= p_{1} \cdot (1 - p_{2}) \cdot 0.5^{2} = p_{1} \cdot (1 - p_{2}) \cdot 0.25 \
\mathbb{P}(CSSC) &= p_{1} \cdot (1 - p_{2}) \cdot 0.5^{2} = p_{1} \cdot (1 - p_{2}) \cdot 0.25 \
\mathbb{P}(SCCS) &= (1 - p_{1}) \cdot p_{2} \cdot 0.5^{2} = (1 - p_{1}) \cdot p_{2} \cdot 0.25 \
\mathbb{P}(SCSC) &= (1 - p_{1}) \cdot p_{2} \cdot 0.5^{2} = (1 - p_{1}) \cdot p_{2} \cdot 0.25 \
\mathbb{P}(SSCC) &= (1 - p_{1}) \cdot (1 - p_{2}) \cdot 0.5^{2} = (1 - p_{1}) \cdot (1 - p_{2}) \cdot 0.25 \
\end{align*}

Sumando las probabilidades, obtenemos:

\begin{equation}
\mathbb{P}(C=2, S=2) = \mathbb{P}(CCSS) + \mathbb{P}(CSCS) + \mathbb{P}(CSSC) + \mathbb{P}(SCCS) + \mathbb{P}(SCSC) + \mathbb{P}(SSCC)
\end{equation}

Simplificando la expresión, tenemos:

\begin{equation}
\mathbb{P}(C=2, S=2)(p_{1}, p_{2}) = -0.5p_1 p_2 + 0.25 p_1 + 0.025 p_2 + 0.25
\end{equation}

```python
p1_values = np.linspace(0.1, 0.9, 100)
p2_values = np.linspace(0.1, 0.5, 100)


p_1, p_2 = np.meshgrid(p1_values, p2_values)


def probability(x,y):
    return -0.5*x*y + 0.25*x + 0.25*y + 0.25
```

```python
min_probability = np.inf
max_probability = 0
for i in p1_values:
    for j in p2_values:
        point = probability(i, j)
        if point > max_probability :
            max_probability = point
            maxs = i,j
        if point < min_probability :
            min_probability = point
            mins = i,j 
```

```python
print("Probabilidad mínima:", min_probability, 'en los puntos '+str(mins))
print("Probabilidad máxima:", max_probability, 'en los puntos '+str(maxs))


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(p_1, p_2, probability(p_1,p_2), cmap='coolwarm')
ax.set_xlabel('p1')
ax.set_ylabel('p2')
ax.set_zlabel('Probabilidad')
ax.set_title('Superficie de probabilidad')



ax.view_init(20,85)
plt.show()
```
