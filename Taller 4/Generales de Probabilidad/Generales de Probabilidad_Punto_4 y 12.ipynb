# Punto 4

La formula general para cacular la probabilidad que $n$ personas ($n \leq 365$) tengan fechas diferentes de cumpleaños es:
$$ \prod_{i=0}^{n}{\frac{(365-i)}{365}} $$

```python
def HP_prob(n):
    
    prod=1
    Pn=np.array([])
    for i in range(n):
        
        prod*=(365-i)/365
        Pn=np.append(Pn,prod)
        
    return Pn
```

```python
n=80
```

```python
X=np.arange(0,n)
Y=HP_prob(n)
```

```python
plt.scatter(X,Y, color='red')
plt.plot(X,Y)
```

$g)$ Un gas ideal conformado por N partículas, realiza una expansión isotérmica de un volumen $V1 = V$ a un un volumen $V2 = 2V$ . Calcule el cambio de entrop´ıa y compare
con el resultado anterior. ¿Cómo se relacionan?


La diferencia de energía interna de un sistema termodinámico esta definida por
$$
\Delta U = \Delta Q-\Delta W
$$

y a su vez es igual a 
$$
\Delta U = \frac{3}{2}kT_2-\frac{3}{2}kT_1
$$

Sin embargo, es un proceso isotérmico, es decir, que no hay cambios en la temperatura con respecto al tiempo. Por lo que la expresión quedaría así

$$
0 = \Delta Q- \Delta W       \rightarrow \Delta Q= \Delta W
$$

El trabajo de un sistema termodinámico esta definido por
$$
\Delta Q=\Delta W=p\Delta V
$$

La integral de la entropía esta dada por

$$ 
\Delta S_{1,2}=\int_{1}^{2}\frac{ \delta Q}{T}
$$

Usando las ecuaciones halladas anteriormente, se tiene que

$$ 
\Delta S_{1,2}=\int_{1}^{2}\frac{ p \delta V}{T}
$$

Despejando $p$ de la Ley de gases ideales ($p=\frac{nRT}{V}) obtenemos


$$ 
\Delta S_{1,2}=\int_{1}^{2}\frac{ nRT \delta V}{T}=nR\int_{1}^{2}\frac{  \delta V}{V}
$$

Se resuelve la integral
$$
\Delta S_{1,2}=nR\ln(\frac{V_2}{V_1})
$$

Teniendo en cuenta que $V1 = V$ y  $V2 = 2V$

$$
\Delta S_{1,2}=nR\ln(\frac{2V}{V})=nR\ln(2)
$$

$n$ representa el número de moles, pero $n$ también puede ser expresado como $n=\frac{N}{N_A}$, siendo $N=$Número de partículas y $N_A=$Número de Avogrado

$$
\Delta S_{1,2}=\frac{N}{N_A}R\ln(2)
$$

Sin embargo, $\frac{R}{N_A}=k_B$, siendo $k_B$ la contante de Boltzmann

$$
\Delta S_{1,2}=k_BN\ln(2)
$$
Dando así el mismo resultado del punto $f$

Se puede concluir que en el límite de altas temperaturas, el cambio de entropía durante una expansión isotérmica es igual a la entropía en ese límite. Esto implica que la variación de entropía en una expansión isotérmica es máxima cuando la temperatura es alta.

Esta conclusión es consistente con el concepto de que a altas temperaturas, la energía térmica es alta y las partículas tienen una mayor agitación térmica. Esto lleva a un mayor número de microestados accesibles y, por lo tanto, a un mayor desorden o incertidumbre en el sistema, que se expresa como un aumento en la entropía.
