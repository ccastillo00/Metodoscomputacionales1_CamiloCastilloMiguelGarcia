# Punto 4

La formula general para cacular la probabilidad que $n$ personas ($n \leq 365$) tengan fechas diferentes de cumplea√±os es:
$$ \prod_{i=0}^{n}{\frac{(365-i)}{365}} $$

```python
def HP_prob(n):
    
    prod=1
    Pn=np.array([])
    for i in range(n):
        
        prod*=(365-i)/365
        Pn=np.append(Pn,prod)
        
    return Pn
```

```python
n=80
```

```python
X=np.arange(0,n)
Y=HP_prob(n)
```

```python
plt.scatter(X,Y, color='red')
plt.plot(X,Y)
```

$g)$ Un gas ideal conformado por N part√≠culas, realiza una expansi√≥n isot√©rmica de un volumen $V1 = V$ a un un volumen $V2 = 2V$ . Calcule el cambio de entrop¬¥ƒ±a y compare
con el resultado anterior. ¬øC√≥mo se relacionan?


La diferencia de energ√≠a interna de un sistema termodin√°mico esta definida por
$$
\Delta U = \Delta Q-\Delta W
$$

y a su vez es igual a 
$$
\Delta U = \frac{3}{2}kT_2-\frac{3}{2}kT_1
$$

Sin embargo, es un proceso isot√©rmico, es decir, que no hay cambios en la temperatura con respecto al tiempo. Por lo que la expresi√≥n quedar√≠a as√≠

$$
0 = \Delta Q- \Delta W       \rightarrow \Delta Q= \Delta W
$$

El trabajo de un sistema termodin√°mico esta definido por
$$
\Delta Q=\Delta W=p\Delta V
$$

La integral de la entrop√≠a esta dada por

$$ 
\Delta S_{1,2}=\int_{1}^{2}\frac{ \delta Q}{T}
$$

Usando las ecuaciones halladas anteriormente, se tiene que

$$ 
\Delta S_{1,2}=\int_{1}^{2}\frac{ p \delta V}{T}
$$

Despejando $p$ de la Ley de gases ideales ($p=\frac{nRT}{V}) obtenemos


$$ 
\Delta S_{1,2}=\int_{1}^{2}\frac{ nRT \delta V}{T}=nR\int_{1}^{2}\frac{  \delta V}{V}
$$

Se resuelve la integral
$$
\Delta S_{1,2}=nR\ln(\frac{V_2}{V_1})
$$

Teniendo en cuenta que $V1 = V$ y  $V2 = 2V$

$$
\Delta S_{1,2}=nR\ln(\frac{2V}{V})=nR\ln(2)
$$

$n$ representa el n√∫mero de moles, pero $n$ tambi√©n puede ser expresado como $n=\frac{N}{N_A}$, siendo $N=$N√∫mero de part√≠culas y $N_A=$N√∫mero de Avogrado

$$
\Delta S_{1,2}=\frac{N}{N_A}R\ln(2)
$$

Sin embargo, $\frac{R}{N_A}=k_B$, siendo $k_B$ la contante de Boltzmann

$$
\Delta S_{1,2}=k_BN\ln(2)
$$
Dando as√≠ el mismo resultado del punto $f$

Se puede concluir que en el l√≠mite de altas temperaturas, el cambio de entrop√≠a durante una expansi√≥n isot√©rmica es igual a la entrop√≠a en ese l√≠mite. Esto implica que la variaci√≥n de entrop√≠a en una expansi√≥n isot√©rmica es m√°xima cuando la temperatura es alta.

Esta conclusi√≥n es consistente con el concepto de que a altas temperaturas, la energ√≠a t√©rmica es alta y las part√≠culas tienen una mayor agitaci√≥n t√©rmica. Esto lleva a un mayor n√∫mero de microestados accesibles y, por lo tanto, a un mayor desorden o incertidumbre en el sistema, que se expresa como un aumento en la entrop√≠a.
