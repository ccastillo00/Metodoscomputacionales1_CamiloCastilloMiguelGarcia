# Punto 3

Usando el algoritmo de Metrópolis, realice el muestreo de $N = 1000 $ eventos de una
distribución normal: $\textit{A∼N}(x;\mu = 2, \sigma =0.5)$

```python
def dist_gauss(x,mu, std):
    return (1/(std*(2*np.pi)**(1/2)))*np.exp(-(x-mu)**2/(2*sigma**2))
```

```python
N=1000
mu=2.
x=np.zeros(N)
x[0]=1.01
sigma=0.5
delta=1
acept=0
```

```python
for i in range(N-1):
    y=x[i]+np.random.uniform(-delta,delta)
    
    if np.random.rand()<min(1,dist_gauss(y,mu,sigma)/dist_gauss(x[i],mu,sigma) ):
        x[i+1]=y
        acept+=1
    else:
        x[i+1]=x[i]
print('La aceptación fue del '+str((acept/N)*100)+'%')
```

```python
plt.hist(x,density=True,bins=50)
xs=np.linspace(0,4,1000)
plt.plot(xs,dist_gauss(xs,mu,sigma),'r')
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend(['Distribución Normal', 'Muestra'])
plt.show()
```

# Punto 4

Usando el algoritmo de Metrópolis, realice el muestreo de $N = 1000 $ eventos de una
distribución estándar de Cauchy: $\textit{A∼}f(x;0,1)$

$$
f(x;0,1)=\frac{1}{\pi (1+x^2)}
$$

```python
def cauchy(x):
    return 1/(np.pi*(1+x**2))
```

```python
N = 1000
sigma = 1
muestra = [0]
sample=[0]
for i in range(N):
    
    xc = np.random.normal(muestra[-1], sigma)
    
    if np.random.uniform(0, 1) <= min(1, Cauchy(xc) / Cauchy(muestra[-1])):
        muestra.append(xc)
    else:
        muestra.append(muestra[-1])
```

```python
plt.hist(muestra, bins=50, density=True)
xp = np.linspace(-10, 10, 1000)
plt.plot(xp, cauchy(xp), 'r-', linewidth=2)
plt.xlabel('x')
plt.ylabel('Densidad de probabilidad')
plt.legend(['Distribución de Cauchy', 'Muestra'])
plt.show()
```
